```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 250
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('crime_data.csv')
df
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
df
```

```python
sns.pairplot(df)
```

```python
#Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()

x = sc.fit_transform(df)
x
```

# Hierarchical

```python
from scipy.cluster import hierarchy
```

```python
lk = hierarchy.linkage(x, method='complete')
dendrogram = hierarchy.dendrogram(lk)
```

```python
#Model Building
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
hc = AgglomerativeClustering(n_clusters=2)

ypred = hc.fit_predict(x)
```

```python
#silhouette
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,ypred)
```

```python
df['cluster'] = ypred
df
```

```python
df.groupby('cluster').agg('mean')
```

```python
df[df['cluster'] == 0]
```

```python
df[df['cluster'] == 1]
```

```python
sns.scatterplot(x=df['UrbanPop'],y= df['Murder'], hue=df['cluster'], palette='rainbow')
```

# Kmeans

```python
#Model Building
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans()
yp = km.fit_predict(x)
yp
```

```python
#cluster centre
km.cluster_centers_
```

```python
#Visualize the clusters
```

```python
plt.scatter(x[:,0],x[:,2], c = yp, cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,2], c = 'black', s = 80)
```

```python
#Elbow graph
```

```python
wcss = []

for i in range(1,11):
    km = KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
plt.axvline(x = 4, linestyle = '--', color = 'red')
plt.axvline(x = 6, linestyle = '--', color = 'red')
plt.xlabel('K-Values')
plt.ylabel('WCSS')
plt.title('Elbow Graph')
```

```python
#Final Model
```

```python
km = KMeans(n_clusters=4)
ypred = km.fit_predict(x)
ypred
```

```python
#Visualize the clusters
plt.scatter(x[:,0],x[:,2], c = ypred, cmap='rainbow')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,2], c = 'black', s = 150)
```

```python
#Silhouette score
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,ypred)
```

```python
#Analyzing clusters
df['cluster'] = ypred
df
```

```python
df.groupby('cluster').agg('mean')
```

```python
sns.scatterplot(x=df['UrbanPop'],y= df['Assault'], hue=df['cluster'], palette='rainbow')
```

# DBSCAN

```python
df.boxplot()
```

```python
#Model Building
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=1, min_samples=6)

y = db.fit_predict(x)
y
```

```python
plt.scatter(x[:,0],x[:,1],c = y,cmap='plasma')
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x,y)
```

```python
#Hyperparameter tuning
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors(n_neighbors=5).fit(x)
```

```python
d,index = neigh.kneighbors(x)
```

```python
d
```

```python
distance = np.sort(d[:,1])
distance
```

```python
plt.plot(distance)
plt.xlabel('Distance')
plt.ylabel('EPS')
plt.title('K-Distance Graph')
```

```python
#Evalute the model
```

```python
dbs = DBSCAN(eps=1.3, min_samples=8)

yp = dbs.fit_predict(x)
yp
```

```python
plt.scatter(x[:,0],x[:,2],c = yp, cmap='plasma')
```

```python
silhouette_score(x,yp)
```

```python
#Analyze the clusters
```

```python
df['Cluster'] = yp
df
```

```python
df.groupby('Cluster').agg('mean')
```
